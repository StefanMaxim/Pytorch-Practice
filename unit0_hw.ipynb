```python
#create two rand tensors of size (2,3) on gpu
torch.manual_seed(1234)
size = (2,3)

#device agnostic code
if torch.cuda.is_available():
    device="cuda"
elif torch.backends.mps.is_available():
    device="mps"
else:
    device="cpu"

tensor1 = torch.randn(size=size).to(device)
tensor2 = torch.randn(size=size).to(device)
tensor1,tensor2

```

```python
#multiply the two above tensors

tenmul = torch.matmul(tensor1,tensor2.T)
tenmul
```
